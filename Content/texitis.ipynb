# Linear Algebra

```python
import math
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.dpi'] = 300

x = np.linspace(-4,4)
m = 2
b = 3
y = m*x+b


plt.plot(x,y)
plt.grid(True)
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.title(r"Plot of $y=2x+3$")
plt.scatter([0],[b],color="green")
plt.plot([-2,-1], [-1,-1],color="red",linestyle="--")
plt.plot([-1,-1], [-1,1],color="red",linestyle="--")
plt.axvline(0,color="green",linestyle="--")
plt.axhline(b,color="green",linestyle="--")
plt.annotate('Rise=2', xy=(-0.9, 0.1), color="red")
plt.annotate('Run=1', xy=(-1.8, -1.4), color="red")
plt.annotate('x=0', xy=(0.1, -2.6), color="green")
plt.annotate('y=b', xy=(-2.5, 3.1), color="green")
plt.annotate(r'$y=2x+3$', xy=(-2.7, 0.1), color="#1f77b4")
plt.annotate('y-intercept: (0,b)', xy=(0, 3), xytext=(0.5, 1.5), color="green",
            arrowprops=dict(facecolor='green', shrink=0.05))
plt.annotate(r'$m=\frac{rise}{run}=2$', xy=(-1.5, -1.5), xytext=(-2.4, -2.7), color="red",
            arrowprops=dict(facecolor='red', shrink=0.05))
plt.xlim([-4,4])
plt.ylim([-4,4])
plt.show()
```

```python
x1a = np.linspace(-1,2.5)
x2a = 1-x1a
x1b = x1a
x2b = 2*x1b

plt.plot(x1a,x2a)
plt.plot(x1b,x2b,color="C3")
plt.xlabel(r"$x_1$")
plt.ylabel(r"$x_2$")
plt.annotate(r'$x_1+x_2=1$', xy=(1, 0.2), color="C0",zorder=0)
plt.annotate(r'$2x_1-x_2=0$', xy=(0.7, 3),color="C3",zorder=0)
# Solution: (1/3, 2/3)
plt.plot([-2,1/3],[2/3,2/3],linestyle="--",color="grey")
plt.plot([1/3,1/3],[-3,2/3],linestyle="--",color="grey")
plt.scatter([1/3],[2/3],color="k",zorder=10)
plt.annotate(r'$x_1=1/3$', xy=(0.38, -1), color="grey",zorder=0)
plt.annotate(r'$x_2=2/3$', xy=(-0.8, 0.8), color="grey",zorder=0)
plt.annotate(r'$(x_1,x_2)=\left(1/3,2/3\right)$', xy=(1/3, 2/3), xytext=(-0.7, 2), color="k",
            arrowprops=dict(facecolor='grey', shrink=0.02))
plt.xlim([-1,2.5])
plt.ylim([-2,5])
plt.title("Linear System Graphical Representation")
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [10, 10*9/16] # Fixes dimensions of plot
plt.rcParams['figure.figsize'] = [9, 5]

x1a = np.linspace(-1,2.5)
x2a = 1-x1a
x1b = x1a
x2b = 2*x1b

fig, axs = plt.subplots(1, 3)

axs[0].plot(x1a,x2a)
axs[0].plot(x1b,x2b,color="C1",linestyle="--")
axs[0].axis("tight")
axs[0].annotate(r'$x_1+x_2=1$', xy=(1, 0.2), color="C0",zorder=0)
axs[0].annotate(r'$2x_1-x_2=0$', xy=(0.2, 4),color="C1",zorder=0)
axs[0].set_title("Unique Solution")

x2b = -x1b
axs[1].plot(x1a,x2a)
axs[1].plot(x1b,x2b,color="C1",linestyle="--")
axs[1].axis("tight")
axs[1].annotate(r'$x_1+x_2=1$', xy=(0.5, 0.6), color="C0",zorder=0)
axs[1].annotate(r'$x_1+x_2=0$', xy=(-1, -0.8),color="C1",zorder=0)
axs[1].set_title("No Solution")

x2b = x2a
axs[2].plot(x1a,x2a)
axs[2].plot(x1b,x2b,color="C1",linestyle="--")
axs[2].axis("tight")
axs[2].annotate(r'$x_1+x_2=1$', xy=(-1, 0.2), color="C0",zorder=0)
axs[2].annotate(r'$2x_1+2x_2=2$', xy=(0.5, 1),color="C1",zorder=0)
axs[2].set_title("Infinitely Many Solutions")

plt.show()
```
